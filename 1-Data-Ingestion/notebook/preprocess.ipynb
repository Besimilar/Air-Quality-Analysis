```python
import pandas as pd
from pandas import DataFrame, read_csv
import datetime
from decimal import *
import numpy as np
import math
```

```python
pd.set_option('display.max_columns', None)
filename = "AQDM_1296629600.txt"
data = read_csv(filename)
data.columns = ["Latitude", "Longitude", "Datum", "Horizontal Accuracy", "State Code", "County Code", "Site Num",
                "Parameter Code", "POC", "AQS Parameter Desc", "Date Local", "24 Hour Local", "Date GMT", 
                "24 Hour GMT", "Year GMT", "Day In Year GMT", "Sample Measurement", "Units of Measure", 
                "Sample Duration", "Sample Frequency", "Detection Limit", "Measurement Uncertainty", 
                "Qualifier Description", "Method Type", "Method Description"]
data
```

```python
data = data.drop(data.index[len(data)-1])
data
```

```python
col = list(data.columns)
col
```

```python
# col.remove("AQS Parameter Desc")
# col.remove("Parameter Code")
# col.remove("Sample Measurement")
# col
```

```python
dp = pd.pivot_table(data, index=data.index.values, columns=["AQS Parameter Desc"], values="Sample Measurement")
dp.head()
```

```python
dp.columns = list(dp.columns)
dp.head()
```

```python
data.shape
```

```python
# data = data.drop("AQS Parameter Desc", axis=1)
# data = data.drop("Parameter Code", axis=1)
# data = data.drop("Sample Measurement", axis=1)
# data.shape
```

```python
frames = [data, dp]
result = pd.concat(frames, axis=1)
result
```

```python
result["Date Local"] = pd.to_datetime(result["Date Local"])
result.shape
```

```python
test = result.sort_values(by=["Latitude", "Date Local", "24 Hour Local"]).reset_index(drop=True)
test
# save_name = "cleaned_data.csv"
# test.to_csv(save_name, index = False)
```

```python
final = test.groupby(["Latitude", "Longitude", "Date Local", "24 Hour Local", "Sample Duration"], as_index=False).mean()
final.shape
```

```python
count = 0
for i in final["Acceptable PM2.5 AQI & Speciation Mass"]:
    if math.isnan(i):
        count += 1
print("Nan number of Acceptable PM2.5 AQI & Speciation Mass: " + str(count))

count = 0
for i in final["PM2.5 - Local Conditions"]:
    if math.isnan(i):
        count += 1
print("Nan number of PM2.5 - Local Conditions: " + str(count))
```

```python
final = final.drop("Parameter Code", axis=1)
final = final.drop("POC", axis=1)
final = final.drop("Sample Measurement", axis=1)
final = final.drop("Detection Limit", axis=1)
final = final.drop("Measurement Uncertainty", axis=1)
final = final.drop("PM2.5 - Local Conditions", axis=1)
final.shape
```

```python
save_name = "cleaned_data.csv"
final.to_csv(save_name, index = False)
```
